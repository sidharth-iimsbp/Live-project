Step 1 : (problem identification)
The problem statement has to be SMART.
Specific: The problem statement has to be detailed and specific to the problem one is solving.
Measurable: Are there any metrics one can track so that one can tell if it is successful at the end of the project
Action: What specific actions can one take to solve the problem
Relevant: There are multiple ways to solve a problem but focus should be on the most relevant method
Timebound: One has to add a time constraint to when the problem should be solved


Step 2 : (Data Collection)
It is the process of gathering and measuring information on targeted variables of interest in an organised system, which then allows to answer relevant questions and decide future outcomes.
A few examples are Kaggle,Company database server,government institution etc.
One should Make sure that the data is Relevant and Validated. If the data is not suitable for the problem one is solving,the results will be useless no matter how good the model is. sQUALITY IS KEY


Step 3 : (Data Cleaning)
One can not overlook this step.Around 80% of the time will be spent in cleaning the data.
Cleaning your data will allow for higher-quality information and ultimately lead to a more conclusive and accurate decision
The actions below lead to a cleaner dataset:
Remove duplicate values ( This is usually the case when combining multiple datasets)
Remove irrelevant observations (observations need to be specific to the problem one is solving)
Address missing values (e.g. Imputation techniques, drop features/observations)
Reformat data types (e.g. boolean, numeric, Datetime)
Filter unwanted outliers (if one has a legitimate reason)
Reformat strings (e.g. remove white spaces,mislabeled/misspelt categories)
Validate (does the data make sense? does the data adhere to the defined business rules?)


Step 4 : Exploratory Data Analysis (EDA)
During this step, one wants to understand the data's statistical characteristics, create visualisations, and test hypothesises.
The major 4 types are 
Univariate non-graphical: make observations of the population and understand sample distributions of a single variable. (e.g. the measure of spread,the measure of central tendency, outlier detection)
Univariate graphical: graphical analysis on a single variable. (e.g. Histograms, Boxplots, Stem and leaf)
Multivariate non-graphical: techniques which show the relationship between two or more variables. (e.g. covariance, correlations)
Multivariate graphical: graphically show the relationship between two or more variables. (e.g. bar plots, scatterplots)


Step 5 : (Feature Engineering)
A feature is an attribute of a dataset that is useful to the problem one is solving. If a feature has no impact on the problem one is solving, it is not part of the problem.
It is defined as the process of transforming raw data into features that better represent the underlying problem to the predictive models, resulting in improved model accuracy on unseen data.
The better the features you create and choose for the  predictive models’, the better the results will be. Feature engineering is an art where you decompose or aggregate, raw data to help solve the problem.
This method includes:
-the correlation coefficient between the feature and the target variable (the feature that one is trying to predict)
-co-integration between two time series (for timeseries data)
-predictive models have embedded feature selection methods(e.g. Random Forest, Gradient Boosting Machine)
- Chi-Squared test (between target and numerical variable)
-Recursive Feature Elimination


Step 6: (Modelling)
All machine learning models are classified either as a Supervised or Unsupervised learning problem.
A Supervised problem is where a function maps an input to an output based on input-output pairs. The machine learning model learns from the input-output training data to make predictions on unseen data (test data). An Unsupervised problem is where a model looks for patterns within an unlabelled dataset.
some models that are used in these categories are
- Supervised - Regression: Linear Regression, Multivariate Linear Regression, Support Vector  Regression (SVR), Random Forest, Neural Networks
--Supervised - Classification: Logistic Regression, Support Vector Machines (SVM),Random Forest, Neural Networks, k-Nearest
Neighbor (kNN)
-  Unsupervised: K-means clustering, Principal Component Analysis (PCA), Singular Value Decomposition (SVD)

After building a few machine learning models, the models needs to be trained by tuning the hyperparameters to optimize model performance. The hyperparameters are parameters used to control the learning process and reduce a predefined loss function. By comparing the predefined metrics for each model, an optimal model can be chosen.


Step 7: (Communication)
Lastly, it is essential to communicate the results. This can be done through a presentation, formal report or even a blog post. A few key points to remember:
-to Use relevant visualisations
-to Know the audience
- to Make sure it flows Data science is about communicating the results well.to Do it with passion, to use a storytelling approach and to show the audience why the findings are so interesting.

# fit the model

```python
X = df.drop('Default',axis=1)

Y = df['Default']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=.30, random_state=123)
```

```python
X_train.head()
```

```python
# Scaling the X_train & X_test

from sklearn.preprocessing import StandardScaler

ss = StandardScaler() 

X_train_scaled = pd.DataFrame(ss.fit_transform(X_train) , columns = X_train.columns) 


X_test_scaled = pd.DataFrame(ss.transform(X_test) , columns = X_test.columns) 
```

```python
# Model Train

from sklearn.neural_network import MLPClassifier

nncl = MLPClassifier(max_iter=2000, 
                     verbose = True , 
                     hidden_layer_sizes= (100,100,100), 
                     activation='logistic' , 
                     solver = 'sgd' , 
                     batch_size=100, 
                     learning_rate='invscaling',
                     learning_rate_init = 0.001,
                     tol=.00001) 
```

```python
nncl.fit(X_train_scaled,Y_train)
```

```python
nncl.score(X_train_scaled,Y_train)
```

```python
nncl.score(X_test_scaled,Y_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
y_pred_train = nncl.predict(X_train_scaled)
y_pred_test = nncl.predict(X_test_scaled)

print(classification_report(Y_train,y_pred_train))
print(classification_report(Y_test,y_pred_test))
```

```python
print(confusion_matrix(Y_train,y_pred_train))
print(confusion_matrix(Y_test,y_pred_test))
```
